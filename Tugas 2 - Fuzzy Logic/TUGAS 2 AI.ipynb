#### 3.Fuzzification

Menerjemahkan tahap membership ke dalam rumus fuzzyfication

```python
def fuzzyPendapatanKrisis(a): 
    if(a <= 0.7):
        return 1
    elif((a>0.7) and (a<0.9)):
        return -(a-0.9)/(0.9-0.7)
    else:
        return 0
```

```python
def fuzzyPendapatanRendah(a):
    if((a>0.7) and (a<0.9)):
        return (a-0.7)/(0.9-0.7)
    elif((a>=0.9) and (a<=1.1)):
        return 1
    elif ((a>1.1) and (a<1.2)):
        return -(a-1.2)/(1.2-1.1)
    else:
        return 0
```

```python
def fuzzyPendapatanSedang(a):
    if((a>1.1) and (a<1.2)):
        return (a-1.1)/(1.2-1.1)
    elif ((a>=1.2) and (a<=1.4)):
        return 1
    elif ((a>1.4) and (a<1.6)):
        return -(a-1.6)/(1.6-1.4)
    else:
        return 0
```

```python
def fuzzyPendapatanTinggi(a):
    if(a >= 1.6):
        return 1
    elif ((a>1.4) and (a<1.6)):
        return (a-1.4)/(1.6-1.4)
    else:
        return 0
```

```python
def fuzzyHutangSedikit(b):
    if(b <= 40):
        return 1
    elif ((b > 40) and (b < 50)):
        return -(b-50)/(50-40)
    else:
        return 0
```

```python
def fuzzyHutangBanyak(b):
    if(b >= 50):
        return 1
    elif ((b > 40) and (b < 50)):
        return (b-40)/(50-40)
    else:
        return 0
```

```python
hasilFuzzyPendapatan = []
hasilFuzzyHutang = []
```

```python
def fuzzyPendapatan(a):
    hasilFuzzyPendapatan.append([fuzzyPendapatanKrisis(a),fuzzyPendapatanRendah(a),fuzzyPendapatanSedang(a),fuzzyPendapatanTinggi(a)])
```

```python
def fuzzyHutang(b):
    hasilFuzzyHutang.append([fuzzyHutangSedikit(b),fuzzyHutangBanyak(b)])
```

```python
for i in range(len(a)):
    fuzzyPendapatan(a[i])
```

```python
for i in range(len(b)):
    fuzzyHutang(b[i])
```

```python
for i in range(len(a)):
    rule1 = min(hasilFuzzyPendapatan[i][0],hasilFuzzyHutang[i][0])
    rule2 = min(hasilFuzzyPendapatan[i][1],hasilFuzzyHutang[i][0])
    rule3 = min(hasilFuzzyPendapatan[i][2],hasilFuzzyHutang[i][0])
    rule4 = min(hasilFuzzyPendapatan[i][3],hasilFuzzyHutang[i][0])
    rule5 = min(hasilFuzzyPendapatan[i][0],hasilFuzzyHutang[i][1])
    rule6 = min(hasilFuzzyPendapatan[i][1],hasilFuzzyHutang[i][1])
    rule7 = min(hasilFuzzyPendapatan[i][2],hasilFuzzyHutang[i][1])
    rule8 = min(hasilFuzzyPendapatan[i][3],hasilFuzzyHutang[i][1])
    hasilInferensi.append([max(rule1,rule5,rule6),max(rule2,rule3,rule7),max(rule4,rule8)])
```

#### 6.De-Fuzzyfication

Grafik Diterima

```python
tidak_x = [0, 40, 50]
tidak_y = [1, 1, 0]

mungkin_x = [40, 50, 60, 70]
mungkin_y = [0, 1, 1, 0]

ya_x = [60, 80, 100]
ya_y = [0, 1, 1]

mat.plot(tidak_x, tidak_y, color='r', label='tidak')
mat.plot(mungkin_x, mungkin_y, color='g', label='mungkin')
mat.plot(ya_x, ya_y, color='b', label='ya')

mat.xlabel('Persentase (%)')
mat.ylabel('Frequency')
mat.title('Grafik Membership Diterima')
mat.legend()
mat.show()
```

```python
deFuzzy = []
```

```python
for i in range(len(hasilInferensi)):
    atas = 0
    bawah = 0
    for j in range (7):
        varBebas = randint(1,99)
        if(varBebas<=40):
            atas=atas+0
            bawah=bawah+0
        elif((varBebas>40) and (varBebas<50)):
            atas = atas+(varBebas*hasilInferensi[i][0])
            bawah = bawah+hasilInferensi[i][0]
        elif ((varBebas>=50) and (varBebas<=60)):
            atas = atas+(varBebas*hasilInferensi[i][0])
            bawah = bawah+hasilInferensi[i][0]
        elif ((varBebas>60) and (varBebas<80)):
            atas = atas+(varBebas*hasilInferensi[i][0])+(varBebas*hasilInferensi[i][2])
            bawah = bawah+hasilInferensi[i][0]+hasilInferensi[i][2]
        elif (varBebas>=80):
            atas = atas+(varBebas*hasilInferensi[i][2])
            bawah = bawah+hasilInferensi[i][2]
    if(bawah==0):
        deFuzzy.append([i,0])
    else:
        deFuzzy.append([atas/bawah])
```

```python
penerimaBLT = [x for _,x in sorted(zip(deFuzzy,dataTugas2["No"]),reverse=True)]
```
